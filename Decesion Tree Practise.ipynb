```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
```

```python
#Downloading Data
my_data = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/drug200.csv', delimiter=",")
my_data.head()
```

```python
my_data["Drug"].unique()
```

```python
features = my_data.iloc[:, :-1]
features["Sex"].replace({'F':0, 'M':1}, inplace=True)
features["BP"].replace({'HIGH':0, 'LOW':1, 'NORMAL':2}, inplace=True)
features["Cholesterol"].replace({'HIGH':0, 'NORMAL':1}, inplace=True)

target = my_data["Drug"]

train_x, test_x, train_y, test_y = train_test_split(features, target, 
                                                    random_state=42, test_size=0.2)
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model.fit(train_x, train_y)
```

```python
model.predict(test_x)
```

```python
plt.figure(figsize=(16,10))
plot_tree(model)
plt.show()
```

```python
model.predict(test_x)
```

```python
model.score(test_x, test_y)
```

```python
model.score(train_x, train_y)
```
